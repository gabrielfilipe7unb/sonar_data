#### Libraries

```python
import pandas as pd
import os
```

#### Setup config

```python
ROOT_DIR = "../"

# Where the datasets are
DATA_PATH = os.path.join(ROOT_DIR, "data")
os.makedirs(DATA_PATH, exist_ok=True)
```

#### Dataframe settings

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

#### Data from JSON

```python
df = pd.read_json('../data/sonar-json.json', lines=True)
```

```python
df = df['measures'][0]
```

#### Data processing

```python
metric = []

history1 = []
history2 = []

history1_value = []
history2_value = []

for i in range(len(df)):
    metric.append(df[i]['metric'])
    history1.append(df[i]['history'][0]['date'])
    history2.append(df[i]['history'][1]['date'])
    try:
        history1_value.append(df[i]['history'][0]['value'])
        history2_value.append(df[i]['history'][1]['value'])
    except KeyError as e:
        history1_value.append(None)
        history2_value.append(None)   
```

#### Data processed

```python
sonar_data = pd.DataFrame([metric, history1, history1_value, history2, history2_value],
                          index=['metric', 'history_1', 'value_1', 'history_2', 'value_2']).T
```

##### Columns:

- metric: metrica
- history_1: histórico de versão 1
- value_1: valor coletado no histórico 1
- history_2: histórico de versão 2
- value_2: valor coletado no histórico 2
